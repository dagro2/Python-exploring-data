# 2. Fill in the missing data using Iterative Imputer in sklearn

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

Pick a ist of columns for imputation: Year, Age, Height, Weight. It's wort marking thatYear doesn't have mssing values, but it might be helpful in modeling the other three.

```python
cols_to_impute = ['Year', 'Age', 'Height', 'Weight']
iter_imp = IterativeImputer(min_value=olympics[cols_to_impute].min(), max_value=olympics[cols_to_impute].max())
imputed_cols = iter_imp.fit_transform(olympics[cols_to_impute])

olympics[cols_to_impute] = imputed_cols
```

Fill the missing values in the column Medal with 'NA'

```python
olympics['Medal'] = olympics['Medal'].fillna('NA')
```

Check the result

```python
olympics.isna().sum()
```

# 8. Calculate the first, third quartile and IQR of the column Age

```python
Q1 = olympics['Age'].quantile(0.25)
Q3 = olympics['Age'].quantile(0.75)
IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

When a data set has outliers, variability is often summarized by a statistic called the interquartile range, which is the difference between the first and third quartiles. The first quartile, denoted Q1, is the value in the data set that holds 25% of the values below it. The third quartile, denoted Q3, is the value in the data set that holds 25% of the values above it.

For Age show the lower and upper thresholds for outliers based on IQR

```python
print(f'Low age outlier threshold: {Q1 - 1.5*IQR}')
print(f'High age outlier threshold: {Q3 + 1.5*IQR}')
```

# 9. Determine which sports are popular among athletes of really young age

```python
lower = (olympics['Age'] < (Q1 - 1.5*IQR))
olympics.loc[lower,'Sport']
```

```python
olympics.loc[lower,'Sport'].value_counts()
```

Swimming and Figure skating are most popular among young athletes

# 10. Determine which sports are popular among athletes of older age

```python
upper = (olympics['Age'] > (Q3 + 1.5*IQR))
olympics.loc[upper,'Sport']
```

```python
olympics.loc[upper,'Sport'].value_counts()
```

Shooting, Art competitions, Equestrianism, Sailing and Fencing are most popular among older athletes
